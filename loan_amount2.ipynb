```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

Импортируем необходимые библиотеки и присваиваем им сокращенные названия.

```python
df = pd.read_csv('data_prep.csv', sep=',' )
```

Импортируем файл с данными используя библиотеку пандас.

```python
df
```

```python
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=13)
```

Выделил нужные Х и У, а затем разделил датасет на тестовый (20%) и обучающий (80%).

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler().fit(X_train)
X_train = sc_X.transform(X_train)
X_test = sc_X.transform(X_test)
```

Производим шкалирование наших данных. Рассчитаем  и оценим в переменной sc_X необходимые параметры шкалирования.

```python
df
```

```python
import statsmodels.api as sm
lr = sm.Logit(y_train, X_train).fit()
print(lr.summary2())
```

Строим базовую модель и анализируем  полученный отчет. Можно сделать вывод, что значимых перменных 2. 

```python
X_train = X_train[:,[5,6]]
X_test = X_test[:,[5,6]]
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(random_state = 13).fit(X_train, y_train)
```

Обучаем модель логистической регрессии.

```python
y_pred = lr.predict(X_test)
lr.score(X_test,y_test)
```

Делаем прогноз на тестовой выборке. Уровень качества модели равен 74%, это значит что  74% объектов распознано верно.

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

Строим таблицу сопряженности. Можно сделать вывод, что 26 положительных случая ложно определены как отрицательные. 5 негативных исходов определены как позитивные. Модель верно определила 92 объекта.

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
```

Устанавливаем необходимые пакеты Keras.

```python
cnn = Sequential()
cnn.add(Dense(units = 4,  activation = 'relu', input_dim = 2))
cnn.add(Dense(units = 1, activation = 'sigmoid'))
cnn.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

Строим нейронную сеть прямой передачи сигнала. На входном слое 4 нейрона. relu - функция  позволит учесть наличие линейнносоти в исходных данных. На втором слое выбирем 1 нейрон используя сигмоидальную функцию.

```python
history = cnn.fit(X_train, y_train, batch_size = 8, epochs = 100, validation_data=(X_test, y_test))
```

```python
y_pred = cnn.predict(X_test)
y_pred = (y_pred > 0.6)
```

Строим прогноз на тестовой выборке. Усложнять архитектуру на такой маленькой выборке, смысла нет.

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
```
